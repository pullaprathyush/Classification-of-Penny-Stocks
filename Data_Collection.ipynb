```python
import yfinance as yf
import pandas as pd
#code to get data from yfinance
def get_stock_info(ticker):
    # Create a Ticker object
    stock = yf.Ticker(ticker)

    # Get stock price
    stock_price = stock.history(period='1d')['Close'].iloc[-1]
    # Get market capitalization
    market_cap = stock.info['marketCap']
    # Get historical stock data for volatility calculation
    historical_data = stock.history(period='1y')['Close']
    # Calculate volatility (you may use more sophisticated methods for volatility calculation)
    volatility = historical_data.pct_change().std()
    # Get liquidity (you may use trading volume as a proxy)
    liquidity = stock.history(period='1d')['Volume'].iloc[-1]
    # Get financial ratios
    pe_ratio = stock.info['trailingPE']
    eps = stock.info['trailingEps']
    # Get listing exchange
    listing_exchange = stock.info['exchange']

    return {
        'Stock':ticker,
        'Stock Price': stock_price,
        'Market Cap': market_cap,
        'Volatility': volatility,
        'Liquidity': liquidity,
        'P/E Ratio': pe_ratio,
        'EPS': eps,
        'Listing Exchange': listing_exchange
    }

output = pd.DataFrame()
```
